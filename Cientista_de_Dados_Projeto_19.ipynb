**Efetividade de Duas Estratégias de Ensino**

Avaliação de uma escola: Vai avaliar  a eficácia de duas estratégias de ensino de matemática para alunos do ensino médio. 
**Objetivo:**Determinar se há uma diferença significativa no desempenho médio dos alunos entre as duas estratégias.

# **Hipóteses:**

* Hipótese nula (H0): A média das notas dos alunos na estratégia A é igual à média das notas dos alunos na estratégia B.
* Hipótese alternativa (H1): A média das notas na Estratégia B é maior do que a média das notas na Estratégia A.

# **Dados:**

* Amostra da Estratégia A: Notas de 50 alunos que receberam a Estratégia A.
* Amostra da Estratégia B: Notas de 50 alunos que receberam a Estratégia B.

Usaremos um teste Z para comparar as médias das notas entre as duas amostras.

Se o p-valor do teste Z for menor que um nível de significância pré-determinado (por exemplo, α = 0.05), rejeitamos a hipótese nula e concluímos que há uma diferença significativa nas médias das notas entre as duas estratégias de ensino.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Os dados são criados a seguir:

```python
# Definindo médias e desvios padrão para as notas nas duas estratégias
media_estrategia_A = 70
desvio_padrao_estrategia_A = 10

media_estrategia_B = 75
desvio_padrao_estrategia_B = 12

# Gerando as amostras de notas para cada estratégia de ensino da nossa base
np.random.seed(0)  # Para garantir a reprodutibilidade dos resultados
amostra_estrategia_A = np.random.normal(loc=media_estrategia_A, scale=desvio_padrao_estrategia_A, size=50)
amostra_estrategia_B = np.random.normal(loc=media_estrategia_B, scale=desvio_padrao_estrategia_B, size=50)

print("Notas da Estratégia A:", amostra_estrategia_A[:5])
print("Notas da Estratégia B:", amostra_estrategia_B[:5])
```

## 1) De acordo com as informações analisadas o nosso teste é unilateral á direita, esquerda ou bicaudal? Justifique.




### É tipo de teste unilateral à direita. 
Justificativa: É que a hipótese alternativa, está testando se a média Estratéfica B é maior que a  média estratégica A.
O interesse é  apenas em desvios positivos. Por isso, o teste é unilateral à direita, pois analisamos a cauda superior da distribuição.


## 2) Calculo das médias para as duas amostragens e as variâncias. Retirada de  insights comparando os dados.

```python
# Cálculo das médias e das variâncias das duas amostras

import numpy as np

# Dados simulados
np.random.seed(0)
amostra_estrategia_A = np.random.normal(loc=70, scale=10, size=50)
amostra_estrategia_B = np.random.normal(loc=75, scale=12, size=50)

# Cálculos
media_A = np.mean(amostra_estrategia_A)
variancia_A = np.var(amostra_estrategia_A, ddof=1)

media_B = np.mean(amostra_estrategia_B)
variancia_B = np.var(amostra_estrategia_B, ddof=1)

print(f"Média Estratégia A: {media_A:.2f}")
print(f"Variância Estratégia A: {variancia_A:.2f}")
print(f"Média Estratégia B: {media_B:.2f}")
print(f"Variância Estratégia B: {variancia_B:.2f}")
```

### Insights:

- A Estratégia B apresenta uma média de notas mais alta, indicando melhor desempenho geral dos alunos em comparação à Estratégia A.

- A variância da Estratégia B é menor, sugerindo que os resultados foram mais consistentes entre os alunos.

- Isso pode indicar que a Estratégia B é mais eficaz e mais estável, beneficiando a maioria dos alunos com menos variação no desempenho.

## 3) Impressão dos resultados da estatística do teste Z, p value.  indicação da rejeição ou não da hipótese nula. 
### Usando statsmodels.stats.weightstats.ztest para Teste Z. Essa função simplifica bastante o cálculo do teste Z, evita erros de implementação e já permite escolher facilmente o tipo de cauda (unilateral ou bicaudal)
### Segue Justificativa da resposta.

```python
# Resultados Teste Z
# colocar os dados array...
# O argumento alternative='larger' faz o teste unilateral à direita.
# Se fosse bicaudal: alternative='two-sided'.
from statsmodels.stats.weightstats import ztest
import numpy as np

# Supondo que seus arrays de notas já estejam definidos:
# notas_A = np.array([...])
# notas_B = np.array([...])

# Exemplo completo com os arrays definidos:
# Definindo os arrays com dados simulados
np.random.seed(0)
notas_A = np.random.normal(loc=70, scale=10, size=50)
notas_B = np.random.normal(loc=75, scale=12, size=50)

# Teste Z, unilateral à direita (H1: média do grupo B > média do grupo A)
z_stat, p_value = ztest(notas_B, notas_A, alternative='larger')

print("Estatística Z:", z_stat)
print("p-valor:", p_value)
```

### Não rejeitar a hipótese nula.
**Justificativa:** O p-valor ultrapassa o limite de 5% , mesmo que por pouco.(se o p-valor > α, não rejeitamos H₀.), indicando que a diferença pode ter ocorrido por acaso. Ainda que a média da estratégia B seja maior, a diferença não é estatisticamente significativa, pois ultrapassa o limite de tolerância para rejeição da hipótese nula . Portanto, não há evidências suficientes para afirmar, ou seja, dar certeza que a Estratégia B seja mais eficaz que a Estratégia A. 

## 4) Gráfico da distribuição da estatística do nosso teste Z. Observação do gráfico.

```python
# Gráfico de distribuição de Z

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Estatística Z observada
z_observado = 1.5267

# Nível de significância (teste unilateral à direita)
alpha = 0.06
z_critico = norm.ppf(1 - alpha)

# Eixo x para a curva normal padrão
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

# Plotando a curva normal
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribuição Normal Padrão', color='blue')

# Linha da estatística Z observada
plt.axvline(z_observado, color='green', linestyle='--', linewidth=2, label=f'Estatística Z = {z_observado:.4f}')

# Linha do ponto crítico
plt.axvline(z_critico, color='red', linestyle=':', linewidth=2, label=f'Z crítico = {z_critico:.3f} (α = 0.05)')

# Área da região de rejeição
x_rejeicao = np.linspace(z_critico, 4, 300)
plt.fill_between(x_rejeicao, norm.pdf(x_rejeicao), color='red', alpha=0.3, label='Região de Rejeição')

# Títulos e legendas
plt.title('Distribuição do Teste Z com Região de Rejeição')
plt.xlabel('Valor de Z')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


```

### O que pode ser observado:

- Uma curva normal padrão (azul)

- Uma linha verde indicando Z = 1.5267

- Uma linha vermelha pontilhada indicando o Z crítico ≈ 1.555

- A área vermelha sombreada mostra a região de rejeição (onde rejeitaríamos H₀)

  
Como o valor de Z está antes do ponto crítico, não rejeitamos a hipótese nula. O gráfico reforça que, embora a média da Estratégia B seja maior, a diferença não é estatisticamente significativa ao nível de 5%.


# 2) Visualizando Gráfico do Teste Z com Região Crítica e Área do p-value
Abaixo está um exemplo em que o gráfico mostra:

A curva normal (distribuição do teste Z),
A região de rejeição (a partir do Z crítico) sombreada,
E a área do p-value (a partir da estatística observada) colorida de forma diferente.

## Aqui está o complemento do código para visualização do teste Z, incluindo o preenchimento correto da região de rejeição (vermelha) e da área do p-valor (laranja):

```python
# código para visualização do teste Z,
# incluindo o preenchimento correto da região de rejeição (vermelha) e da área do p-valor (laranja):

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# valores de exemplo
z_stat = 1.85            # resultado do teste z
alpha = 0.05
z_critico = norm.ppf(1-alpha)  # Z crítico para 5%

# Gera a curva normal padrão
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10,6))
plt.plot(x, y, label='Distribuição Normal Padrão')

# Região de rejeição (a partir do z crítico)
plt.fill_between(x, 0, y, where=(x >= z_critico), color='red', alpha=0.5, label='Região de rejeição (α=5%)')

# Região do p-valor (a partir do z observado)
plt.fill_between(x, 0, y, where=(x >= z_stat), color='orange', alpha=0.7, label='p-valor')

# Linhas auxiliares
plt.axvline(z_stat, color='orange', linestyle='--', label='Estatística Z')
plt.axvline(z_critico, color='red', linestyle='--', label='Z crítico')

plt.legend()
plt.title('Distribuição Z com Região de Rejeição e p-valor')
plt.xlabel('Z')
plt.ylabel('Densidade')
plt.show()
```

## Explicação:
- A área vermelha (à direita de z_critico) mostra onde rejeitamos H0 com α=5%.
    
- A área laranja, a partir de z_stat, mostra visualmente o p-valor: quanto mais à direita ela estiver, menor o p-valor.

- Se z_stat ≥ z_critico, caímos na região de rejeição e rejeitamos H0 com base nos dados.

## Interpretação do Gráfico: Distribuição Z com Região de Rejeição e p-valor
O gráfico mostra a distribuição normal padrão utilizada no teste Z, com destaque para os seguintes elementos:

- Curva azul: Representa a distribuição normal padrão (média 0, desvio padrão 1).

- Linha vermelha tracejada: Indica o Z crítico (≈ 1,645), que é o limite da região de rejeição para um teste unilateral à direita com α = 5%.

- Área vermelha sombreada: Representa a região de rejeição — se a estatística Z estiver dentro dessa área, rejeitamos a hipótese nula.

- Linha laranja tracejada: Marca a estatística Z observada (Z = 1,85).

- Área laranja sombreada: Representa o p-valor, ou seja, a probabilidade de obter um valor de Z igual ou maior que o observado, assumindo que a hipótese nula é verdadeira.

## Conclusão com base no gráfico:
#### Como a estatística Z (1,85) está dentro da região de rejeição, ou seja, à direita do Z crítico, rejeitamos a hipótese nula. Isso indica que há evidência estatística suficiente para afirmar que a Estratégia B é mais eficaz que a Estratégia A ao nível de significância de 5%
